```python
import pandas as pd
import numpy as np
import glob
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
import pickle
import joblib
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import joblib
```

```python

data = []
for f in glob.glob(r'C:\Users\mgomez\OneDrive - dimar.mil.co\Documentos\proyecto_SIGENAV\datos\objetivo_1\recap\datos/*.csv'):
    data.append(pd.read_csv(f, delimiter=';',decimal=','))
data = pd.concat(data)
```

# data science

```python
data.isnull().mean() * 100
```

# ML

```python
df = data[['hs_c_w','wind_speed', 'precipitation','Marea_filtro','Calado','Manga','Eslora']]
df = df.dropna()
df
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
model = OneClassSVM(kernel='rbf', nu=0.01, gamma='auto')
model.fit(X_scaled)
```

```python
pipeline = make_pipeline(StandardScaler(),OneClassSVM(kernel='rbf', nu=0.01, gamma='auto'))
pipeline.fit(df)
joblib.dump(pipeline,r'C:\Users\mgomez\OneDrive - dimar.mil.co\Documentos\proyecto_SIGENAV\codigos\objetivo_1/modelo_SVMO_SPAB.pkl')
```

```python
nuevo_barco = pd.DataFrame({
    'hs_c_w': [.4],
    'wind_speed': [1],
    'precipitation': [50],
    'Marea_filtro': [2],
    'Calado': [18],
    'Manga': [40],
    'Eslora': [250],
})

#scaler = StandardScaler()
#scaler.fit(nuevo_barco)
nuevo_buque_scaled = scaler.transform(nuevo_barco)
resultado_svm = model.predict(nuevo_buque_scaled)  # 1 = normal, -1 = anomalía

print("Puede pasar" if resultado_svm[0] == 1 else "Se recomienda que NO pase")
```
